```python
# Compress model inference to .tar
import tarfile
import os.path

def make_tarfile(output_filename, source_dir):
    with tarfile.open(output_filename, "w:gz") as tar:
        tar.add(source_dir, arcname=os.path.basename(source_dir))

pth ='/content/PaddleOCR/output'
make_tarfile('/content/SROIE_kielayoutlmv2_(epoch5).tar',pth)
```
